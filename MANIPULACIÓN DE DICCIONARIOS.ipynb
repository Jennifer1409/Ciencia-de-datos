```python
"""CREAR UN NUEVO DICCIONARIO DESDE LAS CLAVES DE UNA SECUENCIA
Método: dict.fromkeys(secuencia[, valor por defecto])"""

secuencia = ["animal", "lugar", "población"]
diccionario1 = dict.fromkeys(secuencia)
print(diccionario1)

diccionario2 = dict.fromkeys(secuencia, 'valor x defecto')
print(diccionario2)
```

```python
"""CONCATENAR DICCIONARIOS
Método: update(diccionario)"""

diccionario1 = {"animal": "Notocitellus adocetus", "población": 2721}
diccionario2 = {"lugar": "Apatzingán", "estado de conservación ": "Preocupación menor"}
diccionario1.update(diccionario2)
print(diccionario1)
```

```python
"""ESTABLECER UNA CLAVE Y VALOR POR DEFECTO
Método: setdefault(“clave”[, None|valor_por_defecto])
Si la clave no existe, la crea con el valor por defecto. Siempre retorna
el valor para la clave pasada como parámetro."""

registro = {"animal": "Canis lupus baileyi", "estado de conservación": "En peligro"}
clave = registro.setdefault("lugar", "México")
print(clave)

print(registro)

registro2 = registro.copy()
print(registro2)

clave = registro2.setdefault("Género")
clave
print(registro2)

clave = registro2.setdefault("Reino", "Animalia")
print(clave)

print(registro2)
```

```python
"""OBTENER LAS CLAVES Y VALORES DE UN DICCIONARIO
Método: items()"""

diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
for clave, valor in diccionario.items():
 print(clave, ":", valor)
```

```python
"""OBTENER LAS CLAVES DE UN DICCIONARIO
Método: keys()"""

diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
for clave in diccionario.keys():
  print(clave)

#Obtener claves en una lista

diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
claves = list(diccionario.keys())
print(claves)
```

```python
"""OBTENER LOS VALORES DE UN DICCIONARIO
Método: values()"""

diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
for clave in diccionario.values():
  print(clave)

#Obtener valores en una lista
diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
claves = list(diccionario.values())
```

```python
"""OBTENER LA CANTIDAD DE ELEMENTOS DE UN DICCIONARIO
Para contar los elementos de un diccionario, al igual que con las listas y tuplas,
se utiliza la función integrada len()"""

diccionario = {"animal": "Panthera onca", "lugar": "Reserva de la Biosfera de Calakmul", "población": 120}
print(len(diccionario))
```
